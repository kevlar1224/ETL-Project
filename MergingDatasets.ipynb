```python
import pandas as pd
```

```python
#Ingesting two data sets on top Spotify songs of different years, then concatenating them

top2017 = pd.read_csv("Resources/top2017/top2017.csv", encoding="latin1")
top2018 = pd.read_csv("Resources/top2018/top2018.csv", encoding="latin1")

spotify_df = pd.concat([top2017, top2018])
spotify_df.head()
```

```python
#Getting rid of the extra columns, as we only want to see how the songs did on the Billboard Top 100

spotify_df = spotify_df[['name', 'artists', 'duration_ms']]
spotify_df = spotify_df.rename(columns = {"name":"title", "artists":"artist"})
spotify_df.head()
```

```python
billboard_df = pd.read_csv("Resources/billboard_2000_2018_spotify_lyrics.csv", encoding="latin1")
```

```python
billboard_df.columns
```

```python
#Doing the same for the Billboard dataset

billboard_df = billboard_df[["date", "title", "artist", "peak_pos", "last_pos", "weeks", "rank", "change", "duration_ms"]]
billboard_df.head()
```

```python
#Getting rid of all the songs from before 2017

billboard_df.loc[billboard_df['date'] == "12/31/16"]
```

```python
bb_df = billboard_df.iloc[0:1043]
```

```python
bb_df.tail()
```

```python
#Creating a function that lets us search the "title" and "artist" columns to examine their formatting
#As the Spotify dataset put features in the title column, while the Billboard dataset put them in the artists column,
#we made a way to easily find formatting differences between the datasets

def find(string, df, col):
    find_list = []
    for i in range(len(df)):
        if string.lower() in str(df[col].iloc[i]).lower():
            find_list.append(i)
    for i in find_list:
        print(df[col].iloc[i], i)
```

```python
find("feat", spotify_df, "title")
```

```python
find("feat", bb_df, "artist")
```

```python
find("-", bb_df, "title")
```

```python
#Although we decided against merging using the artist names, this is the code we used to clean up the artist columns
#in the datasets to make them consistent with each other

bb_df['artist'] = bb_df['artist'].str.upper()
bb_df['artist'] = bb_df['artist'].str.split("FEAT").str[0]
bb_df['artist'] = bb_df['artist'].str.split(" AND").str[0]

bb_df.head()
```

```python
spotify_df['artist'] = spotify_df['artist'].str.upper()
spotify_df['artist'] = spotify_df['artist'].str.split("FEAT").str[0]
spotify_df['artist'] = spotify_df['artist'].str.split(" AND").str[0]
```

```python
spotify_df.head()
```

```python
#Here, we are doing the same with song titles, as they were one of the columns we merged on
#We made the same formatting changes to the two datasets, to make them as consistent with each other as possible

bb_df['title'] = bb_df['title'].str.upper()
spotify_df['title'] = spotify_df['title'].str.upper()
```

```python
spotify_df['title'] = spotify_df['title'].str.split(" - ").str[0]
spotify_df['title'] = spotify_df['title'].str.split("(").str[0]
bb_df['title'] = bb_df['title'].str.split("(").str[0]
```

```python
#We used this code to get rid of stray spaces left at then end of song titles after the previous code

for i in range(len(spotify_df)):
    if spotify_df['title'].iloc[i][-1] == " ":
        spotify_df['title'].iloc[i] = spotify_df['title'].iloc[i][:-1]
```

```python
for i in range(len(bb_df)):
    if bb_df['title'].iloc[i][-1] == " ":
        bb_df['title'].iloc[i] = bb_df['title'].iloc[i][:-1]
```

```python
#Now we made the duration columns consistent with each other. The Billboard dataset contained strings, "unknown",
#where it was missing the data for duration. We converted these to None's to drop them, as there was no reliable way to merge them

bb_df['duration_ms'] = bb_df['duration_ms'].replace(to_replace = {"unknown":None})
bb_df = bb_df.dropna(axis=0, subset = ['duration_ms'])
```

```python
#We converted the duration columns to integers, to make sure were the same datatype

import numpy as np
spotify_df['duration_ms'] = spotify_df['duration_ms'].astype(np.int64)
```

```python
bb_df['duration_ms'] = bb_df['duration_ms'].astype(np.int64)
```

```python
#We merged the rows using title and duration

merged_df = spotify_df.merge(bb_df, how="outer", on=["title", "duration_ms"])
merged_df.head()
```

```python
#Then we dropped any rows that didn't have information from the Spotify dataset,
#as these were songs that weren't among the top Spotify songs

merged_df = merged_df.dropna(axis=0, subset=['artist_x'])
```
